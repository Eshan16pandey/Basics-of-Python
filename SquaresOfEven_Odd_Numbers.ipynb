```python
def main():
    # Define the range
    start = 250
    end = 500

    # Iterate through the range
    for number in range(start, end + 1):
        # Check if the number is even
        if number % 2 == 0:
            # Calculate the square of the even number
            square = number ** 2
            # Print the result
            print(f"The square of {number} is {square}")

if __name__ == "__main__":
    main()
```
